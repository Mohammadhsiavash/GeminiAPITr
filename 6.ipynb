```python
import google.generativeai as genai
import os
from dotenv import load_dotenv, find_dotenv
import time
load_dotenv(find_dotenv(), override=True)
os.environ.get('GOOGLE_API_KEY')

genai.configure(api_key=os.environ.get('GOOGLE_API_KEY'))
model = genai.GenerativeModel('learnlm-1.5-pro-experimental')
chat = model.start_chat(history=[])
```

```python
image = genai.upload_file(path='benji.jpg', display_name='benji')
```

```python
print(f'Uploaded file: {image.display_name} as {image.uri}')
```

```python
file = genai.get_file(name=image.name)
```

```python
response = model.generate_content(['Describe the breed', image])
Markdown(response.text)
```
